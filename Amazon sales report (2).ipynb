# Amazon Sales Report 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv(r'C:\Users\kanis\Downloads\Amazon Sale Report.csv', encoding= 'unicode_escape')
#to avoid encoding error, 'unicode_escape'
```

```python
# Sales Overview
total_sales = df[df['Status'] != 'Cancelled']['Amount'].sum()
total_orders = df[df['Status'] != 'Cancelled'].shape[0]
cancelled_orders = df[df['Status'] == 'Cancelled'].shape[0]

print(f"Total Sales: INR {total_sales:.2f}")
print(f"Total Orders: {total_orders}")
print(f"Cancelled Orders: {cancelled_orders}")
```

```python
# Product Analysis
category_sales = df[df['Status'] != 'Cancelled'].groupby('Category')['Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
category_sales.plot(kind='bar')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales (INR)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('sales_by_category.png')
```

### Actionable Insights from the Sales by Category Graph

**Prioritize High Sales Categories:**
Focus on T-shirts, which have the highest sales.

**Enhance Mid-Tier Categories:**

**Investigate Low Performers:**

**Leverage Market Trends:**
Use customer feedback to explore new products in top-selling categories.
    
**Allocate Resources Wisely:**
Direct more resources to high-performing categories while managing low performers cautiously.
    
**Seasonal Promotions:**

**Explore Cross-Selling:**
Identify opportunities to cross-sell between high and mid-tier categories to increase transaction value.

```python
# Fulfillment Analysis
fulfilment_counts = df['Fulfilment'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(fulfilment_counts.values, labels=fulfilment_counts.index, autopct='%1.1f%%')
plt.title('Fulfilment Methods')
plt.savefig('Fulfilment_methods.png')
```

```python
# Geographical Analysis
state_sales = df[df['Status'] != 'Cancelled'].groupby('ship-state')['Amount'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
state_sales.plot(kind='bar')
plt.title('Top 10 States by Sales')
plt.xlabel('State')
plt.ylabel('Total Sales (INR)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('top_10_states_by_sales.png')

```
